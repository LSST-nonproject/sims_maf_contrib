Here's an example of how you can run a MAF metric over multiple simlations and check the output

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import healpy as hp

import lsst.sims.maf.db as db
import lsst.sims.maf.utils as utils
import lsst.sims.maf.metrics as metrics
import lsst.sims.maf.slicers as slicers
import lsst.sims.maf.stackers as stackers
import lsst.sims.maf.metricBundles as metricBundles
import lsst.sims.maf.plots as plots

import glob
import os
```

```python
# Let's use glob to make a list of database files. Replace the path with wherever you have your dbs stored
files = glob.glob('/Users/yoachim/git_repos/sims_featureScheduler_runs1.7/footprint_tune/*10yrs.db')
# glob does not put things in alphabetic order for some reason
files.sort()
files
```

```python
run_names = [filename.split('/')[-1].replace('_v1.710yrs.db', '').replace('_v1.7_10yrs.db', '') for filename in files]
```

```python
results = []
# Loop over each file, save the results in a list
for filename, run_name in zip(files, run_names):
    conn = db.OpsimDatabase(filename)
    # Set each run to save results in a new directory
    outDir = run_name
    resultsDb = db.ResultsDb(outDir=outDir)

    bundleList = []
    sql = 'filter="r"'
    metric = metrics.Coaddm5Metric()
    slicer = slicers.HealpixSlicer()
    summaryStats = [metrics.MedianMetric()]
    plotDict = {'colorMax': 27.4}
    plotFuncs = [plots.HealpixSkyMap()]
    bundleList.append(metricBundles.MetricBundle(metric,slicer,sql, plotDict=plotDict,
                                                 summaryMetrics=summaryStats,
                                                plotFuncs=plotFuncs, runName=run_name))
    
    bd = metricBundles.makeBundlesDictFromList(bundleList)
    bg = metricBundles.MetricBundleGroup(bd, conn, outDir=outDir, resultsDb=resultsDb)
    bg.runAll()
    bg.plotAll(closefigs=False)

    results.append(bundleList)
```

```python
all_summary = [result[0].summaryValues['Median'] for result in results]
```

```python
x = np.arange(np.size(all_summary))
plt.plot(all_summary, 'ko')
plt.ylabel('Median depth in r (mags)')
ack = plt.xticks(x, run_names, rotation=90)
```

```python
# We can also read in the results after the fact
import sqlite3
import pandas as pd

restored_summary = []
for directory in run_names:
    conn = sqlite3.connect(os.path.join(directory, 'resultsDb_sqlite.db'))
    df = pd.read_sql('select summaryValue from summarystats where summaryName="Median"', conn)
    restored_summary.append(np.max(df.values))
```

```python
plt.plot(restored_summary, 'ko')
plt.ylabel('Median depth in r (mags)')
ack = plt.xticks(x, run_names, rotation=90)
```

```python
# We could look at the difference in depths across the sky
for result, name in zip(results[1:], run_names[1:]):
    diff = result[0].metricValues - results[0][0].metricValues
    hp.mollview(diff, title=name +'-' + run_names[0], unit='delta mag')
```
