This is an ipython notebook to postprocess the galaxy counts metric data for various dither patterns and produce a combined plot of the power spectra. The postprocessing done here was used to generate plots presented by Eric Gawiser at the DESC meeting in Feb 2015. 

The dithering patterns used are part of sims_maf, and are generally based on dither patterns presented by Awan, Gawiser, Kurczynski, and Carroll in https://guidebook.com/guide/28858/event/10156541/. 

The raw metric data input here can be generated by running the ditherDriver.py config file in sims_maf/examples/driverConfigs. 

--------

Import the python and MAF items we need. 

```python
import numpy as np
import matplotlib.pyplot as plt
import healpy as hp
%matplotlib inline
import lsst.sims.maf.sliceMetrics as sliceMetrics
import lsst.sims.maf.db as db
import lsst.sims.maf.metrics as metrics
```

```python
# grab the filenames of the galaxy count metric output files.
filenames = !ls $outDir/*Galaxy*npz
print filenames
```

```python
# We will read the data into a comparison slice metric and a run slice metric. 
# This is a little clumsy (having to use both) and should go away in future release of MAF. But for now,
#  we need the comparison slice metric to generate the combined power spectra plot, and the run slice metric to generate 
#  the summary statistics like total power. 
sm = sliceMetrics.ComparisonSliceMetric(useResultsDb=False, outDir=outDir)
rm = sliceMetrics.RunSliceMetric(useResultsDb=False, outDir=outDir)
```

Now read the metric data and convert galaxy counts (N) into dN/<N> (i.e. N-<N> / <N>).
Then, because the edges of the dither pattern induce "ringing" into the power spectra, mask out the edges where the dN/N values fall into the lowest 4th percentile (empirically determined cutoff). 

```python
iids = sm.readMetricData(filenames)
iids2 = rm.readMetricData(filenames)
for iid in iids:
    meanN = np.mean(sm.metricValues[iid].compressed())
    sm.metricValues[iid] = (sm.metricValues[iid]-meanN)/meanN
    rm.metricValues[iid] = (rm.metricValues[iid]-meanN)/meanN
    if sm.metadatas[iid].split()[-1] == 'NoDither':
        continue
    else:
        cutlevel = np.percentile(sm.metricValues[iid], 4.)
        mask = np.where(sm.metricValues[iid]<cutlevel)[0]
        sm.metricValues[iid][mask] = hp.UNSEEN
        rm.metricValues[iid][mask] = hp.UNSEEN
```

```python
# plot the dN/N values. 
rm.slicer = rm.slicers[0]
for iid in iids:
    rm.slicer.plotSkyMap(rm.metricValues[iid], title=rm.metadatas[iid], colorMin=-.4, colorMax=0.3, xlabel="dN/<N>")
```

```python
# Calculate the total power, after conversion to dN/N and masking edges.
totalPower = {}
for iid in iids:
    totalPower[iid] = rm.computeSummaryStatistics(iid, metrics.TotalPowerMetric('metricdata'))[0]
iidsx = np.argsort(totalPower.values())
iidsx = iidsx
print iidsx
for iid in iidsx:
    print rm.metadatas[iid].split()[-1], sm.metadatas[iid], totalPower[iid]
```

```python
labels = []
for iid in iidsx:
    labels.append(sm.metadatas[iid].split()[-1] + ' (power) %.4g' %(totalPower[iid]))
    #print iid, labels[iid]
sm.plotPowerSpectra(iidsx, labels=labels, removeDipole=False)
plt.xlim(0, 500)
plt.yscale('linear')
plt.ylim(0, 0.0005)
plt.legend(loc=[0.9, 0.5], fontsize='smaller', fancybox=True)
```

Some checking into the plots/functions we're using.

Just checking the galaxy numbers function. 

```python
import matplotlib.pyplot as plt
%matplotlib inline
import scipy
def _galCount(coaddm5, apparent_mag):
        dn_gal = np.power(10., -3.52) * np.power(10., 0.34*apparent_mag)
        completeness = 0.5*scipy.special.erfc(apparent_mag-coaddm5)
        return dn_gal*completeness

m = np.arange(20, 30, 0.25)
coaddm5 = 27.5
y = _galCount(coaddm5, m)
constant = 41253/np.sum(y)
y = y*constant
plt.plot(m, y)
plt.xlabel('Magnitude')
plt.ylabel('NGalaxies')
```

The last few plots are investigating the functional shape of the galaxies completeness function.

```python
import scipy 
x = np.arange(24, 29.5, 0.1)
maxCompleteness = 1
m5 = 27
rolloverSpeed = 2.0
y = 0.5*(scipy.special.erfc((x-m5)/(rolloverSpeed/2.0)))
plt.plot(x, y)
plt.xlabel('Magnitude')
plt.ylabel('erfc completeness')
```

```python
x = np.arange(24, 29.5, 0.1)
maxCompleteness = 1
m5 = 27
rolloverSpeed = 2.5
y = 0.5*(maxCompleteness - np.tanh((x-m5)/(rolloverSpeed/4.0)))
plt.plot(x, y)
plt.xlabel('Magnitude')
plt.ylabel('tanh completeness')
```
